```python
#Import libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)

#Leer la data
df = pd.read_csv("movies.csv")
```

```python
#Inspecci√≥n de data
df.head()
```

```python
#Reviso si existen nulls 
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#elimino los nulls que existen
df = df.dropna()
df.dtypes
```

```python
#Transformo algunas columnas de float a int
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df.head()
```

```python
#Verifico que las columnas year y released coincidan en la parte del a√±o
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
b=(df['yearcorrect']-df['year'])
b.mean()
#Los a√±os son diferentes, decido tomar el valor de la columna yearcorrect como el valido
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Remuevo los duplicados
df.drop_duplicates()
```

##### Hago mis predicciones respecto a la data y las ganancias (gross)
###### Alta correlaci√≥n con budget
###### Alta correlaci√≥n con company

```python
#Busqueda de outliers
df.boxplot(column=['gross'])
```

```python
#Scatter plot with budget y gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget VS GrosssEarnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
#Plot de Budget vs Gross usando Seaborn
sns.regplot(x="budget", y="gross", data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#Valores num√©ricos de correlaci√≥n, se usa m√©todo de Pearson(default)
df.corr()
```

```python
#Matriz de correlaci√≥n -> 1: Alta correlaci√≥n
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
## Correlaci√≥n Gross con Company

# asigno un n√∫mero a los atributos que son String
df_numerized=df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
#Matriz de correlaci√≥n -> 1: Alta correlaci√≥n
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()

#Votes y Budget tienen la mayor correlaci√≥n con las Ganancias
#Company no tiene una alta correlaci√≥n
```
